```python
# import panda library
import pandas as pd
```

```python
# Read csv file using read_csv function 
cars = pd.read_csv("mtcars.csv")
```

```python
cars
```

```python
# Find the car with best mpg
cars[cars["mpg"] == cars["mpg"].max()]
```

```python
# Find the car with worst mpg
cars[cars["mpg"] == cars["mpg"].min()]
```

```python
# Find the car with best horsepower
cars[cars["hp"] == cars["hp"].max()]
```

```python
# Draw 5 number summary of displacement
cars.boxplot("disp")
```

```python
# Find the median of horsepower
cars["hp"].median()
```

```python
# Draw historgam for miles per gallon
cars.hist("mpg")
```

```python
# Draw boxplot of miles per gallon for each cylinder type
cars.boxplot("mpg","cyl")
```

```python
# Average mpg for automatic cars
cars[cars["am"] == 1]["mpg"].mean()
```

```python
# Average mpg for manual cars
cars[cars["am"] == 0]["mpg"].mean()
```

```python
# Average of mpg for automatic vs manual, another way
cars.groupby(by=["am"])["mpg"].mean()
```
